```python
def karatsuba(x, y):
    if x < 10 and y < 10:
        return x * y
    else:
        n_digits = len(str(max(x, y)))
        split = n_digits // 2

        a = x // 10 ** split
        b = x % 10 ** split
        c = y // 10**split
        d = y % 10**split

        termino_rojo = karatsuba(a, c)
        termino_azul = karatsuba(b, d)
        termino_verde = karatsuba((a + b), (c + d))

        return termino_rojo * 10 ** (2*split) + \
                (termino_verde - termino_rojo - termino_azul) * 10 **split + \
                termino_azul
```
